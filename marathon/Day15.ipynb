# D15 pandas Split-Apply-Combine Strategy

## Split-Apply-Combine

* Split：將大的數據集拆成可獨立計算的小數據集(拆成男生、女生資料)
    * pd.groupby("columns")
* Apply：獨立計算各個小數據集(成績取平均)
    * pd.groupby("columns").mean()
    * pd.groupby("columns").agg("mean")
* Combine：將小數據集運算結果合併

#### 問題：

運用下列分數資料分析 

1. 找出全年級各科成績最高分與最低分？
2. 找出數學班平均最高的班級？
3. 分析全校女生與男生國文平均差幾分？

```python
# 題目
import pandas as pd
score_df = pd.DataFrame([[1,50,80,70,'boy',1], 
              [2,60,45,50,'boy',2],
              [3,98,43,55,'boy',1],
              [4,70,69,89,'boy',2],
              [5,56,79,60,'girl',1],
              [6,60,68,55,'girl',2],
              [7,45,70,77,'girl',1],
              [8,55,77,76,'girl',2],
              [9,25,57,60,'girl',1],
              [10,88,40,43,'girl',3],
              [11,25,60,45,'boy',3],
              [12,80,60,23,'boy',3],
              [13,20,90,66,'girl',3],
              [14,50,50,50,'girl',3],
              [15,89,67,77,'girl',3]],columns=['student_id','math_score','english_score','chinese_score','sex','class'])
score_df
```

```python
# 找出全年級各科成績最高分與最低分
for i in ["math_score","english_score","chinese_score"]:
    print(i+"'s max,min score:")
    print(score_df[i].max())
    print(score_df[i].min())
```

```python
# 找出數學班平均最高的班級
score_df.drop("student_id",axis=1).groupby("class").mean()
```

```python
# 分析全校女生與男生國文平均差幾分
score_df.drop(["student_id","class"],axis=1).groupby("sex").mean()
```
